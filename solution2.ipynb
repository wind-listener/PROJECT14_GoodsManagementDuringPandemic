### 绘制长春市地图

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
#中文正常显示
plt.rcParams['font.sans-serif']=['SimHei']
```

```python
# plt.figure(dpi=800)
file_path = "附件3：长春市9个区交通网络数据和主要小区相关数据.xlsx"
data3_jd = pd.read_excel(file_path,sheet_name='交通路口节点数据')
data3_xl = pd.read_excel(file_path,sheet_name='交通路口路线数据')
data3_xq = pd.read_excel(file_path,sheet_name='各区主要小区数据')

data3_jd.set_index(data3_jd.columns[0],inplace=True)
data3_xq.set_index(data3_xq.columns[0],inplace=True)
plt.scatter(data3_jd['路口横坐标'],data3_jd['路口纵坐标'],
            s=0.05, c='blue', marker='.', label='道路节点')
plt.scatter(data3_xq['小区横坐标'],data3_xq['小区纵坐标'],
            s=0.1, c='red', marker='x', label='小区')

# 添加标签和标题
# plt.xlabel('横坐标')
# plt.ylabel('纵坐标')
plt.legend()
title = '长春市交通路口节点'
plt.title(title)
plt.savefig('figures\\'+title+'.png', dpi=800)
# 显示图形
plt.show()
```
